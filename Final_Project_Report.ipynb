The three models we considered were Ridge Regression, Random Forest Regression, and Gradient Boosting Tree Regression. This gave us a sense of if the model was overfitting and how it would generalize. If the mean training and cross-validation scores over the folds were significantly different, then we knew that the model was overfitting the data and we should adjust parameters accordingly. The mean cross-validation MAE gave us an estimate of how the model might perform on new data.

After feature selection and parameter tuning, we selected our best model based on mean cross-validation MAE. Then, we ran the model on the holdout test set to evaluate how it will generalize to new data, and if the model should be deployed for use.

For the Ridge model, we used a sweep parameter algorithm which takes a range of hyperparameter options and the algorithm selects where the cross-validation error is the lowest. As for the random forest model, we used the same algorithm to choose the hyperparameters and paid attention to the cross-validation error improvement compared to the training error reductions. For the gradient boosting model, we used a different algorithm in order to save computing power and time due to it having a more complex, computationally expensive algorithm. We used randomized search and grid search with cross-validation, a priocess by which you give the algorithm a range for the hyperparameters and it randomly selects one, using cross-validation, to find a likely fit for the best hyperparameter. For this, we simply ran the algorithm and then compared the baseline train, cross-validation, and standard deviation errors with the same scores of the hypertuned model. Ultimately, the gradient boosting model was the best predictor in terms of overall MAE so we chose this as our final model. This intuitively made sense, and previously we mentioned that for this project a tree-based model would likely be necessary to capture the complex relationships of the data instead of a linear-based model.

### 5.1 Model Performance (6 points)

Briefly interpret what your metrics mean in practical terms.

Include:
- key metrics (e.g., RMSE, $R^2$)

- comparison of training vs validation/test performance

**5.1 Answer:**

Mean Absolute Error (MAE) was used as the key scoring metric in the model building process. MAE calculates the mean across all samples of the absolute value of the difference between the actual and predicted target value for each sample. It provides the average error on a given prediction, and is in the same units as the target, making it an easy scoring metric to interpret in context. The mean MAE score gave us a value for general model performance, while the standard deviation MAE score indicated the relative stability of the model. Since repeated 5-fold cross-validation was used, we were able to calculate the mean training MAE score and compare that to the mean cross-validation MAE score. If the scores were significantly different, then this was a flag that our model was overfitting the training data and we should adjust parameters (e.g. max tree depth, etc.) accordingly.

For our three baseline models, the Ridge model had the worst MAE of `$236,831.33` and was the least stable with a standard deviation of `$47,231.01`, which is about 20% of the mean. The Random Forest model performed best by MAE, producing the lowest score at `$184,979.41`. However, the mean training MAE for the Random Forest model was `$69,031.64`, which is significantly lower than the validation score, and this large difference in training and validation scores showed that the model was overfitting the training data. The Gradient Boosting model yielded a similar baseline MAE to the Random Forest model of `$185,165.87` but has a training score of `$147,718.04`, which was much closer to the validation score and an indication that the model will ultimately generalize better. Both the Random Forest and Gradient Boosting baseline models showed good stability, as they had standard deviations around 6% of the mean.

None of the models showed significant changes in MAE score when the engineered features were added. The largest improvement was the Ridge model, which produced an MAE of `$235,017.13` when the polynomial features were added, while the greatest loss in MAE was the Random Forest model which had an MAE of `$186,078.84` with all of the engineered features.

Backward feature selection resulted in significant improvements in the Ridge model, as the MAE went down to `$209,578.98`. Both tree-based models showed modest improvements as well, with MAE scores of `$183,502.22` and `$183,613.14` for the Random Forest and Gradient Boosting models respectively. These results were not unexpected, as tree models perform an internal form of feature selection during training by splitting on the most important features first and ignoring irrelevant features, so manual feature selection is not overly impactful.

Finally, after using parameter sweep techniques on the Ridge and Random Forest models and both random search and grid search on the Gradient Boosting model, we achieved the following MAE scores:
-   Ridge Model Mean CV MAE Score: `$209,114.46`
-   Random Forest Model Mean CV MAE Score: `$184,802.94`
-   Gradient Boosting Model Mean CV MAE Score: `$179,989.33`

In practical terms, mean MAE training performance has to do with how much is the error (absolute terms) between the actual data points and the model's predictions in USD. This indicates that, on average, how the model's predictions deviate from the actual values in the subset of the data we provided to the model. For validation, we used cross-validation 5-fold in order to better represent what our predictions are doing. This process involves splitting the data into five subsets, training the model on four, and validating it on the fifth, and rotating this process five times in order to get a more accurate picture of what the errors will be like. The best CV MAE Score achieved by any of our models was `$179,989.33`, which is high. This likely means that our model was either not hypertuned enough, or that we have features still in the data we provided the model that creates noise confusing the model's predictions. Finally, we ran the MAE on the test set, a subset of the data we kept out of way during the model building process. This technique will show whether or not our model performs well on new data. The test MAE score was `$226,669.53`. This means we cannot be very confident at all in our model since the errors are so high. In order to improve these errors, we need to look back at how we cleaned the original dataset (removing outlier rows, deleting useless columns, etc.) and to more thoroughly go through the hyperparameter tuning process to ensure we reach a confident error score that results in more accurate predictions.
