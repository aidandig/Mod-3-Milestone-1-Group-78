Several features were missing over 90\% of their values, and we examined each of these features in turn before proceeding. For example, the square footage of the basement had only 50 total non-null values. We concluded that this variable was incomplete and useless, as we would expect a much larger number of properties in the data to have a basement. Some features with high null counts like the flag for if the property includes a hot tub have a null-value in most cases, but a Y in the rare event that there is a hot tub. Here, we elected to retain the feature, and encode the null-values with 0 and the Y values with 1. Overall, we determined that the majority of features with 90\% or more null values would be useless in predicting the overall assessed tax value of each property. However, some of the flag features were retained and encoded with 0 for the null values, as these variables could be useful in the model building process.

Problematic samples were identified as observations with over 55\% of the data missing, so over half the features had a null value for that given sample. This threshold resulted in 281 samples being dropped. In addition, we identified one missing target value, and that sample was also not included.

To address the issue of outliers in the target variable, we used a lower quantile bound of 0.005 and an upper quantile bound of 0.995 to effectively eliminate the bottom and top 0.5% of samples. In total, 774 rows, or about 1\% of the data was dropped using this technique. This histogram displaying the assessed tax value distribution, showed that the distribution was still right-skewed with the majority of property values inside of \$500,000, but some of the significant outliers were removed. In hindsight, if we removed additional outliers and focussed our analysis on, say properties valued under \$1 million, our model would likely have performed much better, and it is reasonable to think that perhaps different models should be used to evaluate standard versus luxury properties.

Our analysis also turned up several highly redundant features. These redundancies were apparent both in the variable descriptions and in observing the relationships between features in the correlation matrix. For example, multiple columns all provided a count of the number of bathrooms, while another group of features all measured the square footage of certain aspects of the finished living space. In these instances, we selected one or two from the group to move forward with and dropped the others to reduce noise, clutter, and redundancy.

A key decision we made was our approach to imputing the remaining missing feature values. We elected a semi-supervised learning strategy and used an iterative imputer with a tree regressor to predict and impute the missing values. Since many of the variables in the housing data are not linear, we felt that to ensure the best accuracy in our imputed values it made sense to use the iterative imputer and tree regressor rather than simply filling these values with the column mean or median.

In general, our data cleaning and preprocessing strategies made the data usable and less noisy, and likely improved the performance of the model. However, there were some missed opportunities, and in the future we would work to scale the usable data down further. We only considered outliers in the target variable, and we removed the lower and upper 0.5\% of all samples based on the target. However, we should have examined and removed outliers within each feature as well. The number of samples in our data is large, and we would still have plenty of data if we had removed outliers across features and not just focused on outliers in the target. Limiting feature outliers and narrowing the range of assessed tax value by focussing our analysis on smaller subset of property types (e.g. homes under \$1 million) would undoubtedly have improved our overall model performance.

### 4.4 Model Selection

Describe how you selected your model.

Include:
- models you tried and why
- how you evaluated generalization
- hyperparameter tuning
- how you chose the final model

**4.4 Answer:**

The three models we considered were Ridge Regression, Random Forest Regression, and Gradient Boosting Tree Regression. While we suspected the tree-based models would generally perform better given the non-linear relationships in the data, we wanted to include one linear model for comparison. We selected the Ridge model to help protect our results against the potential multicollinearity in the features that we observed in the correlation matrix. We selected the Random Forest model given its ability to handle large datasets, non-linear relationships in the data, and different data types with ease. We did not calculate feature importance scores, but if we had, the Random Forest model would also provide more interpretable results of which property characteristics and amenities have the greatest impact on the assessed tax value as compared to other non-linear models. Finally, we selected the Gradient Boosting Tree model thinking that through its sequential tree building process, it might eliminate some error and perform better than the simpler Random Forest model.

To evaluate model generalization we used a train-test split, taking 80\% of the data for training and leaving 20\% for a final evaluation of the model. Throughout the training process, we used repeated cross-validation with five folds and five repeats, and we were able to take the mean training Mean Absolute Error (MAE) score and mean cross-validation MAE score for each model. This gave us a sense of if the model was overfitting and how it would generalize. If the mean training and cross-validation scores over the folds were significantly different, then we knew that the model was overfitting the data and we should adjust parameters accordingly. The mean cross-validation MAE gave us an estimate of how the model might perform on new data.

After feature selection and parameter tuning, we selected our best model based on mean cross-validation MAE. Then, we ran the model on the holdout test set to evaluate how it will generalize to new data, and if the model should be deployed for use.

For the Ridge model, we used a sweep parameter algorithm which takes a range of hyperparameter options and the algorithm selects where the cross-validation error is the lowest. This helped us update a dictionary of hyperparameters for the final model testing. There were times where we had to adjust the range on the parameter values when the gain in cross-validation error was minimial but we started to see the training error signficantly decreasing, which can be an indication of the model overfitting. This is because we do not want the model to memorize the data, we want it to make accurate predictions that will generalize once new data is given. As for the random forest model, we used the same algorithm to choose the hyperparameters and paid attention to the cross-validation error improvement compared to the training error reductions. For the gradient boosting model, we used a different algorithm in order to save computing power and time due to it having a more complex, computationally expensive algorithm. We used randomized search and grid search with cross-validation, a priocess by which you give the algorithm a range for the hyperparameters and it randomly selects one, using cross-validation, to find a likely fit for the best hyperparameter. For this, we simply ran the algorithm and then compared the baseline train, cross-validation, and standard deviation errors with the same scores of the hypertuned model. Ultimately, the gradient boosting model was the best predictor in terms of overall MAE so we chose this as our final model. This intuitively made sense, and previously we mentioned that for this project a tree-based model would likely be necessary to capture the complex relationships of the data instead of a linear-based model.
