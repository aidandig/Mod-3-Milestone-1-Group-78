### 4.1 Analytical Framework

Describe your overall approach.

Include:
- your overall framework
- use of validation curves
- choice of MAE or RMSE (as appropriate) as the primary error metric

**4.1 Answer:**  
Our overall framework was to use models that would best capture the information of our dataset. In doing so, we did not rely on linear models because the relationships between features were more complex than that. Instead, we chose to focus on tree-based models which captures the interactions between features rather than creating a simple line for our predictions (though you can use linear regression forquadratic relationships, trees are more efficient at capturing these relationships). When we graphed the housing data, we could immediately tell that this project has to do with non-linear relationships due to the shape of the distributions. See below:

![image](imgs/zillow_dataset_distribution.png)

The methodology we had was to establish a linear model as the "base model" so we could compare it to more complex models (tree based) in order to see if we were heading in the right direction of more accurately predicting tax assessment evaluations. 

In the model picking phase, for our validation we simply used the mean and standard deviation errors in USD to determine which model was performing better. Our choice of our primary error metric was the mean absolute error because it provides the most direct interpretation of the model's performance in the real-world financial context. The errors are expressed simply in plain dollars rather than square root dollars. Also, MAE ensures that outliers are not disproportionately penalized. We then compared those baseline numbers with new errors when adding our own created features (e.g., feature interactions or polynomial features). During the model tuning phase we used validation curves in order to determine what the best parameters would be using a set range of parameters and comparing the cross-validation MAE and the training MAE. The sweep parameter algorithm finds the sweet spot where the CV MAE is the lowest. We made sure that the improvement was justified without the model continuing to "memorize" the data (the train MAE keeps going down but the CV MAE either goes down an insignificant amount or does not go down at all). By monitoring this, we ensured that we found the best parameters for our model while restricting the model from overfitting the data too much. See below for some validation curve examples:

![image](imgs/validation%20curves%20example.png)
![image](imgs/validation%20curves%20example%202.png)

Note that we did not use validation curve visuals for every model during the model selection phase. This is because we used multiple algorithms that worked best based on the models. For example, the gradient boosting model took significantly longer to sweep parameters so instead we used a randomized search with cross validation to essentially find the best parameters but in a more ranomized fashion rather than simply finding the low point in the CV MAE like in the random forest model with the images above.
